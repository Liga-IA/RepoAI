## Understanding the vectors in grafic

```python
from sklearn.datasets import make_blobs #make_blob purpose is to generate synthetic, multi-class datasets that are particularly useful for testing and demonstrating clustering algorithms.
```

```python
X, y = make_blobs(centers = 4)
```

```python
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0], X[:,1], c=y) #this function signature is matplotlib.pyplot.scatter(x-coordinate, y-coordinate, s=None, c=None)
                                  #and c controls the color of each marker;
                                  #so in this case use the array y to decide the color of each point.
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X,y) #this time we didnt use train_test_split function
```

Lets analise the support vector in our dataset

```python
support_vectors = model.support_vectors_
```

```python
plt.ion()
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(support_vectors[:,0], support_vectors[:,1], c='r')
plt.draw()
```

Now lets import a real dataset

```python
import pandas
```

```python
data= pandas.read_csv('/content/sample_data/diabetes.csv')
```

```python
data.head(3)
```

In this dataset who is X and y? Lets organize the in variables

```python
X = data.iloc[:,0:8].values #all lines and column 0 till 8
y = data.iloc[:,8].values #all lines and column 8
```

```python
X.shape, y.shape
```

Lets separate our data in train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

# SVM support vector machines algorithm

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='linear')
```

```python
svm.fit(X_train, y_train)
```

### Confusion Matrix

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(svm, classes=['0','1'])
```

```python
# To create the ConfusionMatrix, we need some test data. Score runs predict() on the data
# and then creates the confusion_matrix from scikit-learn.
cm.score(X_test, y_test)
cm.show()
```

```python
cm = ConfusionMatrix(svm, classes=['0','1'])
cm.fit(X_train, y_train)
cm.score(X_test, y_test)
cm.poof()
```
