# Example Gradient Boosting

In this example, we will work with the wine classification dataset, which is built into Scikit-Learn

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine() # Initializing the wine data
wine # Full description of the dataset
```

```python
print(wine.feature_names) # Printing the features
print(wine.target_names) # Printing the labels
```

#


# Separating the data into features and target

```python
X = wine.data # Feature
y = wine.target # Target

# Printing the first 5 rows
print(X[0:5])
print('-------')
print(y[0:5])
```


# Separating the data into training and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2, # Separating 20% of the data for the test set
                                                    random_state=42 # For replicating results
                                                    )
```

# Importing Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier # Because we are classifying wines into 0, 1, and 2

gb = GradientBoostingClassifier(n_estimators=100, # number of decision trees
                                learning_rate=0.1, # Learning rate
                                max_depth=1, # Tree depth
                                random_state=42
                                ).fit(X_train, y_train)
```


# Making predictions

```python
y_pred = gb.predict(X_test)
```

# Evaluating the score

```python
gb.score(X_test, y_test)
```

### Although this may sound wonderful, we have to be skeptical and consider whether the model simply memorized the data, leading to overfitting

### Let's analyze the Confusion Matrix and the classification report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
print(classification_report(y_test, y_pred, target_names=wine.target_names))
```

Indeed, the model got every prediction right, correctly classifying all the test data

# Performing Cross-Validation to confirm overfitting and achieve a more reliable model performance

```python
from sklearn.model_selection import cross_val_score
```

We divided the dataset into 3 chunks, training and testing the model multiple times, using different chunks for training and testing for each iteration, retrieving the mean score for each fold

```python

cross_val_score(gb, X_train, y_train, cv=3, n_jobs=-1).mean()
```

# Searching for the best hyperparameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [10, 50, 100, 500],
    'learning_rate': [0.0001, 0.001, 0.01, 1.0],
    'max_depth': [3, 5, 7, 9]
}
```

# Combining each parameter from the object

```python
gb2 = GridSearchCV(gb, param_grid, cv=3, n_jobs=-1)
```

```python
gb2.fit(X_train, y_train)
```

```python
gb2.best_params_ # The best combination of hyperparameters
```

```python
gb2.best_score_ # Best score
```

```python
best_gb_model = gb2.best_estimator_

y_pred_best = best_gb_model.predict(X_test)

cm_best = confusion_matrix(y_test, y_pred_best)
```

```python
print(best_gb_model.score(X_test, y_test)) # best final score, in the test set
print(cm_best)
```

# Conclusion
The inherent datasets in Scikit-Learn are well-behaved; that is, any model will provide an excellent result, as our 1.0 score indicates. The mean score with cross-validation for the 3-chunk division (0.964) shows that the model didn't necessarily memorize all the data, since it used different subsets for testing and training.

In the search for the best hyperparameters, the best score (0.972) reinforces the thesis that the model generalized well for new data, without overfitting.
